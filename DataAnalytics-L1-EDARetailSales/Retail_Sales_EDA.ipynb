Analyzing monthly and quarterly sales trends to identify patterns in retail sales over time.

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
df["Month"] = df["Date"].dt.month_name()
df["Quarter"] = df["Date"].dt.quarter
```

```python
monthly_sales = df.groupby("Month")["Total Amount"].sum()

month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly_sales = monthly_sales.reindex(month_order)

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
quarterly_sales = df.groupby("Quarter")["Total Amount"].sum()

plt.figure(figsize=(6,4))
plt.plot(quarterly_sales.index, quarterly_sales.values, marker='o')
plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

Analyzing customer demographics based on age groups and gender distribution.

```python
bins = [0,  18, 25, 35, 45, 55, 65 ]
labels = ["0-18", "19-25", "26-35", "36-45", "46-55", "56-65"]

df["Age Group"] = pd.cut(df["Age"], bins=bins, labels=labels )
```

```python
plt.figure(figsize=(10,5))
sns.countplot( x="Age Group", data=df )
plt.title("Customer Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["Gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Gender Distribution ")

plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,6))

correlation = df.select_dtypes(include=['number']).corr()

sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```

Customers in the 31–40 and 41–50 age groups contribute significantly to total sales. Businesses can target these age groups with personalized offers and loyalty programs.

```python
df["Age Group"] = pd.cut(
    df["Age"],
    bins=[0,20,30,40,50,100],
    labels=["<20","21-30","31-40","41-50","50+"]
)

age_sales = df.groupby("Age Group", observed=False)["Total Amount"].sum()

plt.figure(figsize=(8,5))
age_sales.plot(kind="bar", color="orange")

plt.title("Total Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Sales")

plt.show()
```
